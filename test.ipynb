```python
import torch 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
def load_data (path, feature_name) : 
    data = torch.load(path) 
    return data[feature_name] 


def plot_feature (path_list, feature_name, legend_list, color_list) : 
    data_list = []
    for path in path_list : 
        data = load_data(path, feature_name) 
        data_list.append(data) 

    plt.figure(figsize=(6, 4))

    for idx, (data, legend, color) in enumerate(zip(data_list, legend_list, color_list)) : 
        print(legend) 
        plt.plot(data, color=color, label=legend) 
        plt.legend() 
        plt.xlabel("Evaluation Step") 
        plt.ylabel(feature_name)
```

```python
path_list = ["./test/ARC_statistics.pt", "./test/NNM_statistics.pt"]
colors = ['#A8D8B9', '#87CEEB'] 
labels = ['ARC', 'NNM'] 

plot_feature(path_list, 'Accuracy', labels, colors) 
```

```python
path_list = ["./test/gas_statistics.pt", "./test/krum_statistics.pt"]
colors = ['#A8D8B9', '#87CEEB'] 
labels = ['gas', 'krum'] 

plot_feature(path_list, 'Accuracy', labels, colors) 
```

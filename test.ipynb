```python
import torch 
import matplotlib.pyplot as plt 
import pandas as pd 
from os.path import join 
```

```python
def load_data (path, feature_name) : 
    data = torch.load(path) 
    return data[feature_name] 


def plot_feature (path_list, feature_name, legend_list, color_list, save_path, ymin=0, ymax=100) : 
    data_list = []
    for path in path_list : 
        data = load_data(path, feature_name) 
        data_list.append(data) 

    plt.figure(figsize=(6, 4))

    for idx, (data, legend, color) in enumerate(zip(data_list, legend_list, color_list)) : 
        print(legend) 
        plt.plot(data, color=color, label=legend) 
        plt.legend() 
        plt.xlabel("Evaluation Step") 
        plt.ylabel(feature_name)
        plt.ylim((ymin, ymax))
    
    plt.savefig(save_path)
```

```python
colors = ['#87CEEB', '#B47ECF',  '#ECD266', '#F4A460', '#A8D8B9'] 
save_path = "./test_ARC_NNM/plots/ALIE_CwTM_ARC_NNM_CrossEntropy_MNIST_4_0.3_20_32_0.9_distrib2_statistics.png"

path_list = [f"./test_ARC_NNM/ALIE_CwTM_{f}_CrossEntropy_MNIST_4_0.3_20_32_0.9_distrib2_statistics.pt" for f in ['ARC', 'NNM']]
plot_feature(path_list, 'Accuracy', ['ARC', 'NNM'], colors, save_path, 95) 
```
